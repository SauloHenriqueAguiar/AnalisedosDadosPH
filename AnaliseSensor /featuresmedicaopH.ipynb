```python
import pandas as pd

# Carregando o DataFrame a partir do arquivo CSV renomeado
df = pd.read_csv("/content/dadospH.csv")

# Dicionário de mapeamento dos nomes antigos para os novos
novo_nome_colunas = {
    'pH(Meio_Aquoso)': 'pHmetodoEmbrapa',
    'pH(Medicao_Direta_Akso)': 'pHsensorComercial',
    'Média-FS(ociloscópio)(mV)': 'OciloscópioMedia-FS',
    'CC-RMS(Ociloscópio)(mV)': 'OciloscópioCC-RMS',
    'Multimetro(DDP)(mV)': 'Multimetro',
    'ESP-32(Média Movel)': 'SensorDesenvolvidoMM',
    'ESP-32(TrueRMS)': 'sensordesenvolvidoTrueRMS'
}

# Renomeando as colunas usando o método `rename()`
df.rename(columns=novo_nome_colunas, inplace=True)

```

```python
# Exibindo o DataFrame após a renomeação das colunas
print(df)
```

```python
import matplotlib as plt


#Correlação de Pearson
correlation = df.corr()
# plot da matriz de correlação
import seaborn as sn
plot = sn.heatmap(correlation,cmap="YlGnBu", annot = True, fmt=".2g", linewidths=.6)

plot
```

```python
# Criando o DataFrame com os dados fornecidos
df = pd.DataFrame(df)

```

```python
# Plotando o scatter matrix usando a biblioteca seaborn
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks")
sns.pairplot(df)

# Exibindo o gráfico
plt.show()
```
