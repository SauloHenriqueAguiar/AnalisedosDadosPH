```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 12, 5
#df = pd.read_excel(file_location).ffill()
sns.lineplot(x=df['ddp_pH_4'].astype(str),
             y=df['T (°C)'],color='r')
sns.lineplot(x=df['ddp_pH_7'].astype(str),
             y=df['T (°C)'],color='g')
sns.lineplot(x=df['ddp_pH_7(b)'].astype(str),
             y=df['T (°C)'],color='b')
sns.lineplot(x=df['ddp_pH_10'].astype(str),
             y=df['T (°C)'],color='y')
plt.legend(['pH4','pH7','pH7b','pH10'])
plt.ylabel('temperatura')
plt.xlabel('ddp')
plt.tick_params(axis='x',labelsize=8)
```
