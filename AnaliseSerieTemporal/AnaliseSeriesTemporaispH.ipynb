```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

```

```python
json_file = open("dadosbancosemana.json")
data_dict = json.load(json_file)
```

```python
# Criar listas para cada campo no DataFrame
time_list = []
humidity_list = []
temperature_list = []
ph_list = []
ph_pred_list = []
dia_list = []
mes_list = []
prev_pH_list = []
```

```python
for key, value in data_dict.items():
    data = json.loads(value)
    time_list.append(data['Time'])
    humidity_list.append(data['Humidity'])
    temperature_list.append(data['Temperature'])
    ph_list.append(data['pH'])
    ph_pred_list.append(data['pHpred'])
    dia_list.append(data['Dia'])
    mes_list.append(data['Mes'])
    prev_pH_list.append(data['PrevpHDiariaSemanal'])
```

```python
# Criar DataFrame com as listas extraídas
data = {
    'Time': time_list,
    'Humidity': humidity_list,
    'Temperature': temperature_list,
    'pH': ph_list,
    'pHpred': ph_pred_list,
    'Dia': dia_list,
    'Mês': mes_list,
    'PrevpHDiariaSemanal': prev_pH_list
}
```

```python
df = pd.DataFrame(data)

```

```python
# Excluindo linhas que contenham pelo menos um valor nulo
df = df.dropna()
```

```python
# Excluindo colunas que contenham pelo menos um valor nulo
df = df.dropna(axis=1)
```

```python
# Excluindo linhas que contenham apenas valores nulos em todas as colunas
df = df.dropna(how='all')
```

```python
# Excluindo linhas que tenham menos de 'thresh' valores não nulos
# Por exemplo, para manter apenas linhas com pelo menos 5 valores não nulos:
df = df.dropna(thresh=5)
```

```python
# Imprimir o DataFrame resultante
print(df)
```

```python
print(df['PrevpHDiariaSemanal'])
```

```python
# Convertendo a coluna 'Time' para o tipo datetime
df['Time'] = pd.to_datetime(df['Time'])

# Definindo 'Time' como índice para manipulação de séries temporais
df.set_index('Time', inplace=True)

```

```python
# Resumindo os dados diariamente
daily_summary = df.resample('D').agg({
    'pH': ['mean', 'std', 'min', 'max'],
    'pHpred': ['mean', 'std', 'min', 'max']
})
```

```python
# Plotando os valores de pH e pHpred ao longo do tempo
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['pH'], label='pH')
plt.plot(df.index, df['pHpred'], label='pHpred')
plt.xlabel('Data')
plt.ylabel('Valor de pH')
plt.title('Valores de pH e pHpred ao longo do tempo')
plt.legend()
plt.grid()
plt.show()
```

```python
# Calculando as métricas
mae = mean_absolute_error(df['pH'], df['pHpred'])
mse = mean_squared_error(df['pH'], df['pHpred'])
rmse = np.sqrt(mse)
# Calculando o MAPE
mape = np.mean(np.abs((df['pH'] - df['pHpred']) / df['pH'])) * 100
```

```python
# Exibindo as métricas
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```

### Analise pHprevisao semanal

```python
humidity_list = []
temperature_list = []
ph_list = []
ph_pred_list = []
dia_list = []
mes_list = []
prev_pH_list = []





for key, value in data_dict.items():
    data = json.loads(value)
    humidity_list.append(data['Humidity'])
    temperature_list.append(data['Temperature'])
    ph_list.append(data['pH'])
    ph_pred_list.append(data['pHpred'])
    dia_list.append(data['Dia'])
    mes_list.append(data['Mes'])
    prev_pH_list.append(json.loads(data['PrevpHDiariaSemanal']))

data = {
    'Humidity': humidity_list,
    'Temperature': temperature_list,
    'pH': ph_list,
    'pHpred': ph_pred_list,
    'Dia': dia_list,
    'Mês': mes_list,
    'PrevpHDiariaSemanal': prev_pH_list
}

df = pd.DataFrame(data)

# Criando uma figura para o gráfico
plt.figure(figsize=(10, 6))

# Plotando os valores reais de pH em azul
plt.plot(df.index, df['pH'], marker='o', color='blue')

# Plotando as previsões de pH para cada dia da semana em amarelo
for i, day in enumerate(df['Dia']):
    for pred_day, pred_value in df['PrevpHDiariaSemanal'][i].items():
        plt.plot(df.index[i],pred_value, marker='o', color='yellow')

# Configurações do gráfico
plt.xlabel('Número de Medições-Previsões')
plt.ylabel('Valores de pH Medido-Previsto')
plt.title('Comparação entre Valores Reais de pH e Previsões Diaria-Semanal')
plt.legend()

# Exibindo o gráfico
plt.show()
```

```python
df = pd.DataFrame(data)

# Calcular métricas de erro
def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def rmse(y_true, y_pred):
    return np.sqrt(mse(y_true, y_pred))
```

```python
# Criar uma lista para armazenar as métricas de erro individuais
mae_values = []
mape_values = []
mse_values = []
rmse_values = []
```

```python
# Comparar os valores de pH reais e previstos para cada dia da semana
for i, prev_pH in enumerate(prev_pH_list):
    pH_real = np.array(df['pH'])
    pH_pred = np.array(list(prev_pH.values()))

    # Verificar se o tamanho de pH_pred é menor que o de pH_real
    if len(pH_pred) < len(pH_real):
        # Preencher pH_pred com a média dos valores adjacentes
        pH_pred = np.repeat(np.mean(pH_pred), len(pH_real))


     # Calcular as métricas de erro para cada valor de previsão
    mae_values.append(mae(pH_real, pH_pred))
    mape_values.append(mape(pH_real, pH_pred))
    mse_values.append(mse(pH_real, pH_pred))
    rmse_values.append(rmse(pH_real, pH_pred))
```

```python
# Calcular as médias das métricas de erro
mae_value = np.mean(mae_values)
mape_value = np.mean(mape_values)
mse_value = np.mean(mse_values)
rmse_value = np.mean(rmse_values)
```

```python
print("MAE:", mae_value)
print("MAPE:", mape_value)
print("MSE:", mse_value)
print("RMSE:", rmse_value)
```
